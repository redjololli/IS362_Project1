```python
import pandas as pd
#import csv to pandas
flights = pd.read_csv('airlines.csv')
flights
```

## Calculating arrival delay rate (%) based on destinations

```python
delay_rate=flights['Airline_delayed']/(flights['OnTime']+flights['Airline_delayed']) * 100
#add to dataframe
flights['delay_rate']=delay_rate
#group by destination and airline
group_by_airline=flights.groupby(['Airport','Airline'])
group_by_airline.sum()
```

## Calculating mean arrival delay rate (%)

```python
group_by_airline=flights.groupby(['Airline'])
group_by_airline.mean()
```

### Conclusion
- According to this AM West has highest mean delay rate.
